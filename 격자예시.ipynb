```python
import geopandas as gpd

# 100m x 100m 격자 파일명 (Colab에 업로드된 파일명 사용)
grid_path = '2. 100M X 100M 신고이력 매핑용 격자.geojson'
grid_gdf = gpd.read_file(grid_path)

# 격자 데이터 확인
print(grid_gdf.head())
print(grid_gdf.crs)
```

```python
# grid_gdf CRS를 gangnam_boundary CRS에 맞추기
grid_gdf = grid_gdf.to_crs(gangnam_boundary.crs)

# 공간조인: 강남구 Polygon과 격자 데이터 겹치는 부분 선택
gangnam_grids = gpd.sjoin(grid_gdf, gangnam_boundary, how='inner', predicate='intersects')

# 결과 확인
print(gangnam_grids.head())
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
gangnam_grids.plot(ax=ax, facecolor='lightblue', edgecolor='black')
gangnam_boundary.boundary.plot(ax=ax, edgecolor='red', linewidth=2)
ax.set_title('강남구 100m x 100m 격자')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1) 가상 데이터 생성: 0~100 사이 랜덤 값
np.random.seed(42)  # 재현 가능하게
gangnam_grids['fake_value'] = np.random.uniform(0, 100, size=len(gangnam_grids))

# (옵션) 범주화해서 discrete 컬러맵 쓰고 싶다면:
# bins = [0, 20, 40, 60, 80, 100]
# labels = ['Very Low','Low','Medium','High','Very High']
# gangnam_grids['fake_cat'] = pd.cut(gangnam_grids['fake_value'],
#                                   bins=bins,
#                                   labels=labels,
#                                   include_lowest=True)

# 2) choropleth 스타일로 그리기
fig, ax = plt.subplots(figsize=(10, 10))

# 격자에 컬러맵 적용 (continuous)
gangnam_grids.plot(
    column='fake_value',      # 혹은 'fake_cat'
    cmap='OrRd',              # 원하는 Matplotlib 컬러맵
    scheme='quantiles',       # 분포에 따라 자동 구간화 (optional)
    k=5,                       # quantiles 개수: fake_cat 사용 안 하면 무시
    legend=True,
    edgecolor='black',
    linewidth=0.2,
    ax=ax
)

# 3) 경계 덮어쓰기
gangnam_boundary.boundary.plot(
    ax=ax,
    edgecolor='red',
    linewidth=2,
    label='강남구 경계'
)

# 4) 제목·레이아웃
ax.set_title('강남구 100m x 100m 가상 데이터 Choropleth', fontsize=16)
ax.axis('off')
plt.legend()
plt.show()
```

```python
# 1️⃣ 필요한 라이브러리 설치
!pip install osmnx geopandas matplotlib

# 2️⃣ 라이브러리 불러오기
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt

# 3️⃣ 강남구 경계 데이터 가져오기
gangnam_boundary = ox.geocode_to_gdf('강남구, 서울특별시, 대한민국')

# 4️⃣ 강남구 경계 확인 (옵션)
print(gangnam_boundary)
fig, ax = plt.subplots(figsize=(8, 8))
gangnam_boundary.plot(ax=ax, facecolor='none', edgecolor='red')
ax.set_title('강남구 경계')
plt.axis('off')
plt.show()

# 5️⃣ 지구대별 집계정보 데이터 불러오기
file_path = '23. 관서별 지구대별 집계정보.geojson'  # 업로드된 파일명
gdf_police = gpd.read_file(file_path)

# 6️⃣ 지구대별 데이터 좌표계 확인
print(gdf_police.crs)

# 7️⃣ 좌표계 맞추기 (강남구 Polygon 기준으로 변환)
gdf_police = gdf_police.to_crs(gangnam_boundary.crs)

# 8️⃣ 공간조인으로 강남구에 포함된 지구대만 선택
gdf_gangnam_police = gpd.sjoin(
    gdf_police,
    gangnam_boundary,
    how='inner',
    predicate='intersects'
)

# 9️⃣ 결과 확인
print(gdf_gangnam_police[[ '주거지역_집계정보', '상업지역_집계정보', '공업지역_집계정보', '녹지지역_집계정보']])

# 🔟 Choropleth: 주거, 상업, 공업, 녹지 한꺼번에 시각화 (subplot)
cols = ['주거지역_집계정보', '상업지역_집계정보', '공업지역_집계정보', '녹지지역_집계정보']
titles = ['주거지역 신고건수', '상업지역 신고건수', '공업지역 신고건수', '녹지지역 신고건수']

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

for idx, (col, title) in enumerate(zip(cols, titles)):
    ax = axes[idx // 2, idx % 2]
    gdf_gangnam_police.plot(
        column=col,
        cmap='OrRd',
        legend=True,
        edgecolor='black',
        ax=ax
    )
    gangnam_boundary.boundary.plot(ax=ax, edgecolor='red', linewidth=2)
    ax.set_title(f'강남구 지구대별 {title}', fontsize=14)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# 주거, 상업, 공업, 녹지지역 신고건수 합산 컬럼 추가
gdf_gangnam_police['총_신고건수'] = (
    gdf_gangnam_police['주거지역_집계정보'] +
    gdf_gangnam_police['상업지역_집계정보'] +
    gdf_gangnam_police['공업지역_집계정보'] +
    gdf_gangnam_police['녹지지역_집계정보']
)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))
gdf_gangnam_police.plot(
    column='총_신고건수',
    cmap='OrRd',
    legend=True,
    edgecolor='black',
    ax=ax
)
gangnam_boundary.boundary.plot(ax=ax, edgecolor='red', linewidth=2)
ax.set_title('강남구 지구대별 총 신고건수', fontsize=14)
plt.axis('off')
plt.show()
```
